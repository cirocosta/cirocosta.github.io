## Conceitos Básicos

- Relação entre CG e Processamento de Imagens

Processamento de Imagens trata de transformações do domínio das imagens para as imagens ($F:Img->Img$) ou dos domínio das Imagens para as representações/modelos ($F:Img->Modelos$), enquanto a disciplina de CG trata da transformação de modelos bem descritos para imagens ($F:Modelos->Img$) através de um processo análogo aos formados por sistemas de imagem físicos (câmeras, microscópios...). Assim, para compor uma imagem precisamos de: objetos, materiais, observador e fonte(s) de iluminação.



## Renderização

### primitivas de baixo nível disponíveis e de primitiva de alto nível

São disponibilizadas apenas primitivas de baixo nível pelo fato de ser a maneira mais rápida para desenhá-los já que a biblioteca pode partir de certos princípios sabendo que todos os pontos dentro do mesmo são do mesmo plano (qualquer ponto interno pode ser referenciado por coordenadas baricêntricas, variando linearmente, por exemplo). Podendo ter então mais assunções pode-se otimizar tal desenho.

Um exemplo de primitiva de alto nível que seria útil seria uma esfera, que no fim, nada mais é do que um *mesh* de triângulos que pode ser obtido através de subdivisões de um icosaédro, por exemplo.

### Passos para desenhar um objeto tridimensional na tela do seu computador

Os passos que envolvem tal desenho são descritos pelos passos performados na pipeline do OpenGL. **tl;dr:, a aplicação preenche os buffers de memória do OpenGL com arrays de vértices, que são projetados então ao universo do canvas, montados então em triangulos e rasterizados em fragmentos do tamanho de pixels (pixels potenciais). Cada fragmento recebe uma cor e então os aceitos são lançados ao framebuffer que joga na tela.** O OpenGL permite então que, mais do que escrever o código da aplicação que joga os vértices para a GPU, também escrevamos o código a ser utilizado dentro da GPU. Estes últimos tratam-se dos *shaders*.

Primeiramente é necessário que seja construída a representação do modelo, que fixa para si mesmo um centro (sua origem) e define então todos os vértices dele em função desta. Tal representação trata-se de um array de vértices (coordenadas no seu espaço local com referência ao centro) que então será lançado a algum buffer que passará tal informação ao Vertex Shader, possivelmente carregando diversas outras propriedades, tais como coordenadas de texturas, cores, etc. 

1. São convertidos os vértices entre sistemas de representações: $model \to view \to perspective$. É necessário que combine um observador no mundo 3D e objetos deste a fim de então produzir uma imagem 2D para ser desenhada no canvas. Este é o trabalho da última matriz aplicada (a perspectiva), enquanto a segunda (visão) trata de passar do espaço do modelo (centralizado no mesmo) para o espaço do mundo (centralizado na câmera, que enxerga o mundo). O Vertex Shader trata então de passar estas novas coordenadas para a parte de montagem dos triângulos e o FragmentShader.

2. Tendo os vértices transformados é necessário então agrupá-los em objetos geométricos: pontos, segmentos de retas, polígonos.

3. cortam-se os excessos da imagem (pontos onde não haverá visualização) de modo a ter menos trabalho a realizar

4. Para cada objeto gerado (após a montagem) produz-se fragmentos (pixels em potencial) que terão então cores (determinadas por mapeamento de textura ou interpolação de cores de cada vértice) associadas ao tal, assim como posição no frame buffer e também z-buffer (profundidade). É comum que o trabalho de iluminação e aplicação de textura seja feito aqui (no Fragment Shader), onde é trabalhado no nível de pixels.

5. aqueles fragmentos potenciais que são 'bloqueados' por outros fragmentos mais próximos da câmera são eliminados (hidden surface removal) de acordo com posicionamento no z-buffer. Os sobreviventes são então desenhados na tela conforme indicado pelo framebuffer após as filtragens.




## Coordenadas Homogêneas

Coordenadas homogêneas permitem-nos **unificar o tratamento de vetores e pontos**, levando o problema para uma dimensão superior: coordenada extra `w=0` para vetores e `w=1` para pontos (na verdade depende do tipo de projeção: na ortográfica mantemos 0 e 1, enquanto na perspectiva precisamos de um fator de divisão).

Desta forma **todas as tranformações clássicas podem ser implementadas por multiplicações**.

Coordenadas homogêneas surgem do fato de que um sistema de coordenadas é insuficiente para representar pontos. Passamos então para um espaço afim, adicionando um ponto (origem) aos vetores base (agora base = (P0, v0, v1, v2)) para formar um *sistema de referência*, tendo então vetores representados pela forma $v=\alpha_1*v_1 + \alpha_2*v_2 + ... + \alpha_n*v_n$ e pontos como $P = P0 + \beta_1*v_1 + ... + \beta_n*v_n$.

Utilizando então coordenadas homogêneas a representação de vetores nunca levará a adição de um `PO`, enquanto pontos serão definidos como deslocamentos da origem até a posição.

Para alternarmos entre sistemas de referência precisamos então levar em conta os pontos de origem (já que estes fazem parte da definição de base) ao performar a transformação de um para outro. Trata-se exatamente do mesmo processo de mudança de base convencional porém agora, ao definirmos um ponto na nova somamos a origem da antiga base.

- Qual a diferença entre pontos e Vetores?

Enquanto pontos são locais no espaço, vetores são deslocamentos.

Dado um ponto podemos adicionar deslocamento; dado dois pontos podemos inferir o deslocamento para chegar de um ao outro; dados dois deslocamentos podemos somá-los. Entretanto, não podemos adicionar dois locais (não faz sentido).

Vale notar que: vetores apresentam magnitude e direção, sem apresentar posição. Pontos apenas apresentam sua posição no espaço.



## Equações Implícitas e Paramétricas

- Equações Paramétricas

A ideia de ter equações paramétricas para curvas é a de expressar as coordenadas de pontos na curva através de apenas um parâmetro dado (em superfícies, 2 ... etc), tendo a forma:

$$
\begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} g(t) \\ h(t) \end{bmatrix}
$$

ou

$$
p = f(t)
$$

**reta**: uma reta pode ser considerado um conjunto de pontos determinados por um vetor posicionado que então escalamos por um certo valor. A direção de tal vetor é dada pela diferença entre dois pontos (que descrevem a orientação da reta). Podemos tomar a posição do vetor como um dos pontos e então escalar para qualquer valor. Assim: $p(t) = p_0 + t(p_1 - p_0)$.

**circulo**: Sendo $r$ o raio: 

$$
\left[ \begin{array}{c} x \\ y \end{array} \right] = \left[ \begin{array}{c} x_c + r*cos\phi \\ y_c + r*sen\phi \end{array} \right]
$$ 

Podemos entender que trata-se de um deslocamento para o centro de onde o círculo será desenhado e então cada ponto do círculo em si a composição de dois vetores, sendo o vetor horizontal dado pelo cosseno e o vertical pelo seno.

Ps.: Se adicionarmos constantes $a$ e $b$ para os vetores podemos então obter elipses alinhadas aos eixos: $[x | y]^T = [x_c + a*r*cos\phi | y_c + b*r*sen\phi]^T$

ps.: Note que se pensarmos em curvas em 3D continuamos com apenas um parâmetro ($t$) porém operamos em vetores de dimensão 3 ($[x | y | z]^T = [g(t) | h(t) | i(t)]^T$)

**planos**: requer 2 parâmetros ($u,v$).

**esfera**: Enquanto para o círculo as funções para $x$ e $y$ eram dadas por meio do sistema de coordenadas polares (qualque r ponto é determinado por uma distância - no caso, uma constante R - e um ângulo - no parâmetro variável), para a esfera os mesmos são representados a partir do sistema de coordenadas esféricas, que depende então de dois ângulos (nossos dois parâmetros). Sua derivação vem de planos de revolução que dita que a equação paramétrica para qualquer plano de revolução é dado por

$$
(f(u)\cos{v}, f(u)\sin{v}, g(u))
$$

podemos imaginar que a equação paramétrica da esfera derive de uma rotação completa em torno do eixo x da curva de um circulo. Assim:

$$
(r\cos{u}\cos{v}, r\cos{u}\sin{v}, r\sin{u})  =>

x = r \cos{\phi}\sin{\theta}
y = r \sin{\phi}\sin{\theta}
z = r \cos{\theta}
$$

Apesar de sermos 'compelidos' a dizer que as equações paramétricas são a melhor representação de planos em CG, as funções implícitas também possuem grande valia, como na no teste de intersecção e teste inside/outside. Isto se dá pela própria forma de representação: em paramétrica representamos tudo por pontos (fácil de imaginar que para vértices isto cai muito bem), enquanto para implícita definimos planos através de contornos de uma função ($f(p) = 0$).
