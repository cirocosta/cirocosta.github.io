## kNN

1. Dada uma amostra desconhecida, calcula sua distância em relação a cada uma das amostras da base de treinamento
2. seleciona as K amostras mais próximas
3. dentre as selecionadas, escolha a classe que seja maioria. Caso ocorra empate, escolhe a mais próxima entre as que empataram

Para as distância podemos utilizar simplesmente euclidiana ao quadrado $D(u,v) = \sum^{m}_{i=0}(u_i-v_i)^2$.

Podemos implementar tais operações de maneira bastante eficiente utilizando de [broadcasting](http://wiki.scipy.org/EricsBroadcastingDoc), seguindo a regra:
> In order to broadcast, the size of the trailing axes for both arrays in an operation must either be the same size or one of them must be one


```python
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
plt.subplot(2,1,1)

YX = ndimage.imread('../assets/digits_train.png', flatten=True)
X = YX[:,1:] # contains actual matrix of features
Y = YX[:,0]  # labels
plt.title('Training (%d, %d)' % X.shape)
plt.axis('off')
plt.imshow(X, cmap="gray")

plt.subplot(2,1,2)

YXt = ndimage.imread('../assets/digits_test.png', flatten=True)
Xt = YXt[:, 1:]
Yt = YXt[:, 0]
plt.title('Test (%d, %d)' % Xt.shape)
plt.axis('off')
plt.imshow(Xt, cmap="gray")
```

Uma operação básica em 'vector quantization' é a de calcular a distância entre um objeto a ser classificado e múltiplos pontos já classificados (como no nosso exemplo). Utilizando de broadcasting, podemos ter o seguinte exemplo:

```python
observation = np.array([111.0, 188.0])
codes = np.array([[102.0, 203.0],
                  [132.0, 193.0],
                  [45.0, 155.0],
                  [57.0, 173.0]])

diff = codes - observation
print diff
dist = np.sqrt(np.sum(diff**2))
print 'Nearest: %d' % (np.argmin(dist))
```

```python
# shitty
def knn(x_test, Y, X, k=3):
    """kNN implementation using sqr(euclid_dist)
    as distance function.
    
    x_test: sample to be classified
    Y: labels
    X: training
    """
    labels = []
    for i in xrange(x_test.shape[0]):
        diff = np.square(X.astype('float') - x_test[i,:].astype('float'))
        sums = np.sum(diff, axis=1)
        k_closest = np.argsort(sums)
        hist, bins = np.histogram(k_closest[:k])
        prediction = np.argmax(hist)
        labels.append(Y[prediction])
        
    return np.array(labels)

print Yt
print knn(Xt, Y, X, k=7)
```
